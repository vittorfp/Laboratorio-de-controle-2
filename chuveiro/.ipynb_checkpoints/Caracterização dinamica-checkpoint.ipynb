```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.fftpack import fft
from scipy.signal import stft, butter, lfilter, lfiltic
import control as ct
from scipy.optimize import differential_evolution

sns.set_style('whitegrid')
```

```python
file = './dadostorneira/modoCapetinha1'

df = pd.read_csv(file, sep=' ')#, header=None, names=['temp', 'escadinha1', 'c', 'escadinha2', 'time1', 'time2'])
df.pop('CicloFim')
df.pop('CicloInicio')
df.pop('ModoOperacao')
df.plot(figsize=(20, 5));
```

##### Retirada de spikes

Vamos retirar os spikes cujo Z-score local é maior que 3. Definimos Z-score local como sendo o Z-score com relação a media e desvio padrão em uma janela deslizante de 10 amostras. Dessa forma estamos retirando os pontos que acreditamos ser artefatos com 95% de confiança.

```python
window = 10

df['mean_dadofapk'] = df['dadofapk'].rolling(window).mean()
df['std_dadofapk'] = df['dadofapk'].rolling(window).std()

spikes = df.query('(((dadofapk - mean_dadofapk)/std_dadofapk) > 2) or (((dadofapk - mean_dadofapk)/std_dadofapk) < -2)').index
df['dado_sem_spike'] = df['dadobruto'].copy()
df['dado_sem_spike'].loc[spikes] = [(df['dadobruto'].iloc[s+1] + df['dadobruto'].iloc[s-1])/2 for s in spikes]
df[['dadobruto', 'dado_sem_spike']].plot(figsize=(20, 5));
```

## Filtragem do sinal

Vamos projetar um filtro para os ruídos no sinal. Para isso, vamos realizar uma análise no dominio da frequencia, a fim de definir a melhor frequencia de corte, na qual o ruído será retirado mas as propriedades dinâmicas do sistema serão preservadas.

```python
def get_psd_db(signal):
    psd = np.abs(fft(df['dado_sem_spike']))
    psd = psd[:len(psd)//2]
    return 20*np.log10(psd)

def plot_psd_db(signal):
    
    n = len(df['dado_sem_spike'])
    plt.figure(figsize=(21,4))
    plt.plot(np.linspace(0, 0.5, n//2), get_psd_db(signal))
#     plt.yscale('log')
    plt.xscale('log')
```

```python
plot_psd_db(df['dado_sem_spike'].values)
plt.title('Conteúdo espectral (sinal inteiro)');
```

```python
plot_psd_db(df['dado_sem_spike'].iloc[:100].values)
plt.title('Conteúdo espectral (degrau)');
```

```python
plot_psd_db(df['dado_sem_spike'].iloc[300:1500].values)
plt.title('Conteúdo espectral (linha de base)');
```

```python
f, t, Zxx = stft(df['dado_sem_spike'], nperseg=100)
st_psd = np.abs(Zxx)
plt.pcolormesh(t, f, st_psd, vmin=np.min(st_psd), vmax=np.max(st_psd)//700)
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

A julgar pelo conteúdo espectral, a maior parte da potência se encontra nas frequencias menores que 1/10 da frequencia de amostragem, para tal iremos projetar um filtro com corte próximo dessa faixa e avaliar os resultados.

```python
b, a = butter(4, 0.06)
a, b
```

The filter function is implemented as a direct II transposed structure. This means that the filter implements:

a[0]*y[n] = b[0]*x[n] + b[1]*x[n-1] + ... + b[M]*x[n-M]
                      - a[1]*y[n-1] - ... - a[N]*y[n-N]

where M is the degree of the numerator, N is the degree of the denominator, and n is the sample number. It is implemented using the following difference equations (assuming M = N):

a[0]*y[n] = b[0] * x[n]               + d[0][n-1]
  d[0][n] = b[1] * x[n] - a[1] * y[n] + d[1][n-1]
  d[1][n] = b[2] * x[n] - a[2] * y[n] + d[2][n-1]
...
d[N-2][n] = b[N-1]*x[n] - a[N-1]*y[n] + d[N-1][n-1]
d[N-1][n] = b[N] * x[n] - a[N] * y[n]

where d are the state variables.

The rational transfer function describing this filter in the z-transform domain is:

                    -1              -M
        b[0] + b[1]z  + ... + b[M] z
Y(z) = __ ___ ___ __ __ __ ___ ___ ___ __ __ __ ___ __ __ ___ __           X(z)
 
                    -1              -N
        a[0] + a[1]z  + ... + a[N] z


a[0]*y[n] = b[0]*x[n] + b[1]*x[n-1] + ... + b[M]*x[n-M]
                      - a[1]*y[n-1] - ... - a[N]*y[n-N]

```python
y_buffer = [0]*5
x_buffer = [0]*5
filtered = list()
i = [0]*10
i.extend([4000]*100)
for k in i:
    
    y = (
        b[0]*k + b[1]*x_buffer[-1] + b[2]*x_buffer[-2] + b[3]*x_buffer[-3] + b[4]*x_buffer[-4]
        - a[1]*y_buffer[-2] - a[2]*y_buffer[-3] - a[3]*y_buffer[-4] - a[4]*y_buffer[-5]
    )/ a[0]
    
    filtered.append(y)
    
    if len(y_buffer) >= 5:
        y_buffer.pop(0)
    y_buffer.append(y)
        
    
    if len(x_buffer) >= 5:
        x_buffer.pop(0)
    x_buffer.append(k)

plt.plot(filtered)
# df['dado_sem_spike'].plot()
```

```python
N = 4
d = [0]*4
filtered = [0]

for x in df['dado_sem_spike'].values:
    
    y    = ((b[0] * x)              + d[3]) / a[0]
    d[3] =  (b[1] * x) - (a[1] * y) + d[2]
    d[2] =  (b[2] * x) - (a[2] * y) + d[1]
    d[1] =  (b[3] * x) - (a[3] * y) + d[0]
    d[0] =  (b[4] * x) - (a[4] * y)
    filtered.append(y)

df['dado_sem_spike'].plot(figsize=(20, 4))
plt.plot(filtered)
plt.xlim([100, 5000])
plt.ylim([39500, 42500])
```

```python
df['filtered'] = lfilter(a, b, df['dado_sem_spike'])

df[['dado_sem_spike', 'filtered']].plot(figsize=(20, 5))
plt.ylim([39e3, 43e3]);
```

O resultado parece satisfatório para uma frequencia de corte igual a 0.1 da frequencia de 0.05 da frequencia de Nyquist do sinal. A redução no ruído foi satisfatória e o impacto na resposta dinâmica parece aceitavel. Um problema foram as condições iniciais do filtro que não foram calculadas de antemão e portanto houve um artefato no inicio da resposta.

# Modelagem da dinâmica

Para modelar o comportamento veremos quais parâmetros de modelo mais se encaixam aos dados que possuimos.

#### Separação dos degraus

```python
intit_offset = 200
diff = df.iloc[intit_offset:]['ValorDesejado'].diff()
diff.plot();
```

```python
step_times = np.where((diff > 10) | (diff < -10))[0] + intit_offset
steps = []
for st in step_times:
    df_ = df[['filtered', 'ValorDesejado', 'dado_sem_spike']].iloc[st-10:st+1000]
    df_['filtered'] = (df_['filtered'] - df_['filtered'].iloc[0])
    df_['dado_sem_spike'] = (df_['dado_sem_spike'] - df_['dado_sem_spike'].iloc[0])
    df_['ValorDesejado'] = (df_['ValorDesejado'] - df_['ValorDesejado'].iloc[0])
    df_[['filtered', 'ValorDesejado']].plot()
    steps.append(df_)

df_ = steps[-1]
```

### Otimização dos parâmetros do modelo

Para cada um dos degraus vamos obter os parâmetros que melhor descrevem a dinâmica do sistema.

```python
def first_order(tau, gain, delay=20):
    s = ct.tf([1, 0],[0, 1])
    return (gain / (tau*s + 1)) * ct.tf(*ct.pade(delay, n=5))

def second_order(Wn, zeta, gain, delay=20):
    s = ct.tf([1, 0],[0, 1])
    return (gain*Wn**2  / (s**2 + 2*Wn*zeta*s + Wn**2)) * ct.tf(*ct.pade(delay, n=1))

def MSE(signal, simulated_signal):
    return np.mean(np.sqrt(np.square(simulated_signal-signal)))
```

```python
def get_objective(df_):
    
    def objective(args):
        tau, gain, delay = args
        sys = first_order(tau, gain, delay=delay)
        simulated = ct.forced_response(
            sys,
            T=range(len(df_.index)),
            U=df_['ValorDesejado'].values
        )[1]
        return MSE(df_['filtered'], simulated)
    return objective

results = [differential_evolution(get_objective(df), [(0, 1000), (-20, 20), (0, 20)]) for df in steps]
results
```

```python
for _df, res in zip(steps, results):
    _df['simulated_step'] = ct.forced_response(
        first_order(res.x[0], res.x[1], delay=res.x[2]),
        T=range(len(_df.index)),
        U=_df['ValorDesejado'].values
    )[1]
    _df[['filtered', 'simulated_step']].plot(figsize=(15,5));
```

### Ensemble

Os modelos obtidos acima são os melhores para descrever aquele dado no qual eles foram obtidos, porêm eles podem não ser tãso bons em outros conjuntos de dados, para isso, iremos utilizar uma combinação dos 3 modelos obtidos para gerar um modelo mais generalizavel.

Nesse caso utilizaremos a media dos 3 modelos.

```python
mean_model = np.mean(np.array([res.x for res in results]), axis=0)
mean_model
```

```python
for _df in steps:
    _df['simulated_step'] = ct.forced_response(
        first_order(mean_model[0], mean_model[1], delay=mean_model[2]),
        T=range(len(_df.index)),
        U=_df['ValorDesejado'].values
    )[1]
    _df[['dado_sem_spike', 'filtered', 'simulated_step']].plot(figsize=(15,5));
```

### Observações

Os dados acima mostram um comparativo entre o dado bruto, filtrado e modelado. Vemos que os aspéctos essenciais do sistema estão sendo representados, portanto, temos um modelo adequado. 

A utilização do filtro causa um atraso no sinal, porém, como o modelo do sistema foi obtido a partir dos dados filtrados isso não causará problemas, pois o filtro estará na frente do controlador no momento em que o sistema estiver rodando e o controlador será capáz de reagir de forma adequada a esse atraso do filtro.


```python
WWn = np.linspace(0.01, 0.5, 1000)
teta = (np.pi)/2 * WWn

ret = np.exp(1j*teta)*np.sin(teta)/teta
trap = np.tan(teta)/teta

plt.plot(ret)
plt.plot(trap)
plt.plot([1.05 for i in range(1000)])
plt.plot([0.95 for i in range(1000)])
plt.legend(['ret', 'trap']);
```
