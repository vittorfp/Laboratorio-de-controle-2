```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
file = './dadostorneira/dadoscoletados.txt'
# file = './dadostorneira/testeescada1.txt'

df = pd.read_csv(file, sep=' ', header=None, names=['temp', 'escadinha1', 'c', 'escadinha2', 'time1', 'time2'])
df.pop('time1')
df.pop('time2')
df.pop('c')
df.plot(figsize=(30, 10))
```

```python
df_ = df.query('escadinha1 > 40000 and escadinha1 < 60000 and escadinha1 != 45000')
sns.regplot(df_['escadinha1'], df_['temp'])
# plt.ylim([35000, 45000])
```

```python
steps = np.where(df.escadinha1.diff().abs()[100:] > 500)[0] + 100
```

```python
X = [np.mean(df.temp[s-20:s]) for s in steps]
y = [df.escadinha1[s-2] for s in steps]

plt.scatter(y, X);
```
