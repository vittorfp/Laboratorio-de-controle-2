```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('saida.txt', sep=' ', decimal=',').iloc[:, [2, 5, 6, 7, 8]]
df.columns = ['time', 'meassured_level', 'setpoint_level', 'input_valve', 'output_valve']
df.index = list(map(pd.Timestamp, df.pop('time').values))
df.head()
```

```python
df.plot(figsize=(20, 6));
```

```python
step_times = [
    (pd.Timestamp('2019-03-27 16:32:35'), pd.Timestamp('2019-03-27 16:50:35')),
    (pd.Timestamp('2019-03-27 16:57:35'), pd.Timestamp('2019-03-27 17:04:30')),
    (pd.Timestamp('2019-03-27 17:04:30'), pd.Timestamp('2019-03-27 17:12:25')),
    (pd.Timestamp('2019-03-27 17:12:25'), None)
]

def plot_all(df, step_times):
    n_plots = len(step_times)
    fig, ax = plt.subplots(n_plots, 2, figsize=(21, 4 * n_plots))
    for (init, end), a in zip(step_times, ax):
        step = df.loc[init: end]
        step[['meassured_level', 'setpoint_level']].plot(title='Level and setpoint', ax=a)
        step[['input_valve', 'output_valve']].plot(title='Input and output', ax=a)


plot_all(df, step_times)
```
